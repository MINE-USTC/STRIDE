### Minimal index from (title, text) pairs

Replace `pairs` with your corpus. For large corpora, batch titles/texts and call `add_docs` (see `my_retriever.py` in the package).

```python
from contriever_model import load_contriever_and_tokenizer
from my_retriever import DenseRetriever

pairs = [
    ("Example title", "Example paragraph text."),
]

model, tok = load_contriever_and_tokenizer("facebook/contriever")
r = DenseRetriever(model, tok, batch_size=32)
for title, text in pairs:
    r.add_doc(text, title)

out_dir = "faiss_index/mycorpus/index"
r.save_index(out_dir)
print("saved", out_dir)
```
